### Task 1: Conceptual Understanding

#### 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different words if case is not normalized.  
After converting to lowercase, both become the same.

**Example:**  
"Love is beautiful" and "I love coding" → both become "love"

---

#### 2. What happens if stopwords are not removed?
If stopwords are not removed, the text becomes noisy and less efficient for processing.

**Example:**  
"The movie is very good" → "movie good" (after removing stopwords)

---

#### 3. When removing stopwords is harmful

**a) Sentiment Analysis:**  
"I do not like this" → "like" (changes meaning)

**b) Question Answering:**  
"What is your name?" → "name" (loses meaning)

---

#### 4. Stemming vs Lemmatization

- **Stemming:** Cuts words into root form (may be incorrect)  
- **Lemmatization:** Converts words into meaningful base form

**Example:**  
"running" →  
Stemming: "runn"  
Lemmatization: "run"

### Task 2:Build Advanced Preprocessing Function


```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove special characters/punctuation
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Remove very short words (≤2), but keep "no", "not"
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]
    
    return " ".join(words)
```

### Task 3: Stress Testing


```python
# Sample inputs
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Testing loop
for i, text in enumerate(texts, 1):
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    
    print(f"Example {i}:")
    print("Original Text   :", text)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

### Task 4: Token Analytics


```python
# Token Analytics

for i, text in enumerate(texts, 1):
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = round(sum(len(t) for t in tokens) / total_tokens, 2) if total_tokens > 0 else 0
    
    print(f"Example {i}:")
    print("Cleaned Sentence :", cleaned)
    print("Total Tokens     :", total_tokens)
    print("Unique Tokens    :", unique_tokens)
    print("Avg Token Length :", avg_token_length)
    print("-" * 50)
```

### Task 5: Frequency Analysis


```python
from collections import Counter

# Combine all tokens
all_tokens = []

for text in texts:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

# Count frequencies
word_freq = Counter(all_tokens)

# Top 10 most frequent words
top_10 = word_freq.most_common(10)

# Top 5 least frequent words
least_5 = word_freq.most_common()[:-6:-1]

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word}: {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word}: {count}")
```

### Task 6: Build Full Pipeline


```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        cleaned = preprocess_text(text)
        tokens = cleaned.split()
        
        clean_sentences.append(cleaned)
        all_tokens.extend(tokens)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(texts)

print("Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```

### Task 7: Error Handling


```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove emojis & special characters
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Remove short words (≤2), keep "no", "not"
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]
    
    return " ".join(words)
```

```python
edge_cases = [
    "",                # Empty string
    "😍😂🔥",          # Only emojis
    "1234567890",      # Only numbers
]

for text in edge_cases:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    
    print("Original:", repr(text))
    print("Cleaned :", cleaned)
    print("Tokens  :", tokens)
    print("-" * 40)
```

### Edge Case Handling

- Empty string → returns empty output  
- Only emojis → removed as non-alphabetic  
- Only numbers → removed during preprocessing  

This ensures the pipeline does not crash and produces clean outputs.
