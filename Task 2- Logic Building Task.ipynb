##  1)Unique Words in a Sentence


```python
# Take input from user
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list to set to get unique words
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## Highest Salary from Employee Data


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
max_name = ""
max_salary = 0

for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        max_name = name

print(f"Highest Salary: {max_name} - {max_salary}")
```

### Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 100, 66]
max=numbers[0]
min=numbers[0]
for i in numbers:
    if i>max:
        max=i
        print("maximum value",max)


numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

### Count Products Above a Price Threshold


```python
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:
    if i>1000:
        count+=1
        print(i)
print("Products above 1000:",count)        
```

### Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
    if i=="P":
        count+=1
percentage=(count/len(attendance))*100
print("Attendance Percentage: ",percentage)
    
```

### Remove Duplicate Phone Numbers


```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique=set(phone_numbers)
print("Unique phone numbers: ",unique)
```

### Count Character Frequency


```python
text = "python"
frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```
