### 1. User Login Check

```python
# Stored credentials
username = "admin"
password = "1234"

# Taking user input
entered_username = input("Enter username: ")
entered_password = input("Enter password: ")

# Checking login credentials
if entered_username == username and entered_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
fail=0
Pass=0
for i in marks:
    if i>=50:
        Pass+=1
    else:
        fail+=1
print(f"Final Result: Pass:{Pass},Fail:{fail}")
```

### 3. Simple Data Cleaner


```python
# Original list
names = [" Alice ", "bob", " CHARLIE "]

# Cleaning the data
cleaned_names = [name.strip().lower() for name in names]

# Printing cleaned list
print("Cleaned Names:", cleaned_names)
```

### 4. Message Length Analyzer


```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    if length > 10:
        print("Status: Long Message")
    else:
        print("Status: Short Message")
    
    print("-" * 40)
```

### 5. Error Message Detector

```python
# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counter for errors
error_count = 0

# Checking each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1

# Printing total error count
print("Total ERROR entries:", error_count)
```
