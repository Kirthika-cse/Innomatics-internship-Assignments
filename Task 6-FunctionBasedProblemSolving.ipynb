### 1:Smart Parking Lot Management System


```python
def smart_parking(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

### 2: Online Food Delivery Time Estimator

```python
def delivery_time(distance, traffic, weather):
    time = distance * 5   # base time (5 minutes per km)

    # traffic adjustment
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    else:
        time += 5

    # weather adjustment
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

### 3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)

    occupancy = (booked / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1]*150

seat_occupancy(total_seats, booked_seats)
```

### 4: Cloud Server Load Classification System

```python
def server_load(cpu_readings):

    total = 0
    for cpu in cpu_readings:
        total += cpu

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

server_load(cpu_readings)
```

### 5: Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resources):

    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(resources)
```

### 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):

    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
