### 1: Mobile Recharge Validation System

```python
def recharge_validation(amount, plan):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Recharge Rejected: Amount must be at least ₹50"
    
    if plan not in valid_plans:
        return "Recharge Rejected: Invalid plan selected"
    
    return "Recharge Successful"


while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan (199, 299, 399, 599): "))

    result = recharge_validation(amount, plan)
    print(result)

    if result == "Recharge Successful":
        break

    retry = input("Do you want to retry? (yes/no): ").lower()
    if retry != "yes":
        print("Recharge cancelled")
        break
```

### 2: Inventory Reorder Alert System


```python
def inventory_alert(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")


products = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18
}

inventory_alert(products)
```

### 3: Student Result Processing System


```python
def student_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} → Pass"
    else:
        return f"Average: {average:.2f} → Fail"


marks = [60, 45, 70, 50, 55]

result = student_result(marks)
print(result)
```

### 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    total = base_fare + (distance * per_km_rate)

    if peak == "yes":
        total += total * 0.25   # add 25% extra

    return f"Total Fare: ₹{total:.2f}"


while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        result = calculate_fare(distance, peak)
        print(result)

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you!")
            break

    except ValueError:
        print("Invalid input. Please enter numeric distance.")
```

### 5: Employee Attendance Eligibility Checker

```python
def attendance_eligibility(attendance):
    present_days = 0

    # Count present days using loop
    for status in attendance:
        if status == "P":
            present_days += 1

    # Calculate percentage
    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"


attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

result = attendance_eligibility(attendance_list)
print(result)
```
