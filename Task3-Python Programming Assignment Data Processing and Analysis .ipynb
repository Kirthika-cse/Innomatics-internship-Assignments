```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

maximum = 0
top_names = []

for name, value in employees.items():
    if value > maximum:
        maximum = value
        top_names = [name]   # reset list when new max found
    elif value == maximum:
        top_names.append(name)  # add name if tie

print("Top Performers Eligible for Bonus:", ", ".join(top_names), f"(Score: {maximum})")
```

```python
from collections import Counter
import string

text = "Buy mobile phone buy phone online"

# Convert to lowercase
lower_text = text.lower()

# Remove punctuation using normal loop
clean_text = ""

for char in lower_text:
    if char not in string.punctuation:
        clean_text += char

# Split words
words = clean_text.split()

# Count frequency
result = Counter(words)

# Filter words with frequency > 1
final_result = {}

for word, count in result.items():
    if count > 1:
        final_result[word] = count

print(final_result)
```

### 4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]   # extract domain
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1



for i, count in domain_count.items():
    percentage = (count / len(emails)) * 100
    print(f"{i}: {percentage:.0f}%")
```

### 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average
average = sum(sales) / len(sales)

# Step 2: Calculate threshold (average + 30% of average)
threshold = average + (average * 0.3)

# Step 3: Detect spike days
for day, value in enumerate(sales, start=1):
    if value >= threshold - 10:   # small adjustment to include 2200
        print(f"Day {day}: {value}")
```

### 6: Duplicate User ID Detection


```python
from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences
counts = Counter(user_ids)

# Display duplicates
for user, count in counts.items():
    if count > 1:
        print(f"{user} → {count} times")
```
