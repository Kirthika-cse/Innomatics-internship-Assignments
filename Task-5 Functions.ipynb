### Agriculture – Rainfall Adequacy Checker


```python
def calculate_rainfall(rainfall, required_level):
    
    total=sum(rainfall)
    average=total/len(rainfall)
    if average>=required_level:
        print("Average Rainfall:",average)
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Average Rainfall:",average)
        print("Rainfall Status: Inadequate Rainfall")

rainfall=[60,75,80,90,65]
required_level=73
calculate_rainfall(rainfall,required_level)
```

### Social Media – Duplicate Account Detection


```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["john", "anna", "mike", "john"]
check_duplicates(users)
```

### Premium Crop Price Filter

```python
def premium_crops(prices):
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 3200, 1800]
premium_crops(crop_prices)
```

### Application Health Checker

```python
def app_health(errors):
    print("Error Count:", errors)
    
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


app_health(7)
```

### Student Attendance Eligibility

```python
def attendance_check(attendance):
    total_classes = len(attendance)
    present = 0
    
    for status in attendance:
        if status == "P":
            present += 1
    
    percentage = (present / total_classes) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance_list = ["P", "P", "A", "P", "P"]
attendance_check(attendance_list)
```

### Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("Total Bill:", bill)
    print("Usage Level:", usage)


electricity_bill(250)
```
